# Illustration of the Box-Cox Transformation

```python
# Generate random dataset
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Apply Box-Cox transformation on original data
plt.style.use('ggplot')

 # to make reproducible example
np.random.seed(254) 
variable_y = np.random.exponential(size=15_000_000)

# Plot distribution of data
plt.figure(figsize=(7, 7))
plt.title("Histogram of Variable Y")
sns.histplot(variable_y, kde=True)
plt.show()
```

```python
# Apply Box-Cox transformation on original data
plt.style.use('ggplot')
from scipy.stats import boxcox, skew
updated_variable_y, best_lambda = boxcox(variable_y)

print(f'Optimal Box-Cox transformation lambda: {best_lambda:.3f}')

# Plot distributions of original data
plt.figure(figsize=(14, 5))
ax1 = plt.subplot(121)
sns.histplot(variable_y, kde=True, ax=ax1)
plt.title("Histogram of a Right-Skewed $Y$")

# Plot distribution of transformed data
ax2 = plt.subplot(122)
sns.histplot(updated_variable_y, kde=True, ax=ax2)
plt.title(f'Histogram of the transformed updated Variable Y; $\\lambda$ = {best_lambda:.2f}, \n$Skewness$ = {skew(variable_y):.2}')
plt.show()
```
